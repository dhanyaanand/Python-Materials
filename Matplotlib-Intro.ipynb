# Draw a line in a diagram from position (0,0) to position (6,250):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

# Line between (1,3) and (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints,color="red")
plt.show()
```

# Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints,color="green")
plt.show()
```

# Barplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["Ananya", "Bharath", "Cat", "Dana"])
y = np.array([30, 80, 10, 100])

plt.bar(x,y,color="red",width=.1)
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x,color="red")
plt.show() 
```

# pie Chart

```python
import matplotlib.pyplot as plt
import numpy as np
c=["red","blue","green","yellow","black"]
y = np.array([75, 25, 25, 10,90])

plt.pie(y,colors=c)
plt.show() 
```
